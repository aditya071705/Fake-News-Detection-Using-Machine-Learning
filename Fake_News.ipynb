```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
import re
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_fake=pd.read_csv('Fake.csv')
data_true=pd.read_csv('True.csv')
```

```python
data_fake['class']=0
data_true['class']=1
```

```python
data_fake_manual=data_fake.tail(10)
for i in range(23480, 23470, -1):
    data_fake.drop([i],axis=0,inplace=True)

data_true_manual=data_true.tail(10)
for i in range(21416, 21406, -1):
    data_true.drop([i],axis=0,inplace=True)
```

```python
data_merged=pd.concat([data_fake,data_true], axis=0)
data_merged
```

```python
data=data_merged.drop(['title','subject','date'],axis=1)
data
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
data.head(10)
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
data.head(10)
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

print(f"x_train shape: {len(x_train)}")
print(f"y_train shape: {len(y_train)}")
```

```python
print(x_train.dtype)
print(x_test.dtype)
```

```python
x_train=x_train.astype(str)
x_test=x_test.astype(str)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer()
x_vector_train = vector.fit_transform(x_train)
x_vector_test = vector.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(x_vector_train, y_train)
```

```python
predict_LR=logreg.predict(x_vector_test)
Accuracy_Score=accuracy_score(y_test,predict_LR)
print(Accuracy_Score)
```

```python
Classification_Report=classification_report(y_test,predict_LR)
print(Classification_Report)
```

```python
from sklearn.tree import DecisionTreeClassifier

DTC=DecisionTreeClassifier()
DTC.fit(x_vector_train, y_train)
```

```python
predict_DTC=DTC.predict(x_vector_test)
```

```python
Accuracy_Score=accuracy_score(y_test,predict_DTC)
print(Accuracy_Score)
```

```python
Classification_Report=classification_report(y_test,predict_DTC)
print(Classification_Report)
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier()
RFC.fit(x_vector_train, y_train)
```

```python
predict_RFC=RFC.predict(x_vector_test)
```

```python
Accuracy_Score=accuracy_score(y_test,predict_RFC)
Accuracy_Score
```

```python
Classification_Report=classification_report(y_test,predict_RFC)
print(Classification_Report)
```

```python
def output(n):
    if n==0:
        return " It is a Fake News"
    elif n==1:
        return "It is Not A Fake News"
    
def testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_x_vector_test = vector.transform(new_x_test)
    prediction_LR = logreg.predict(new_x_vector_test)
    prediction_DTC = DTC.predict(new_x_vector_test)
    prediction_RFC = RFC.predict(new_x_vector_test)
    
    return print("\n\nLR Predicition: {} \nDTC Prediction: {} \n RFC Prediction:{}".format(output(prediction_LR[0]),
                                                output(prediction_DTC[0]),
                                                output(prediction_RFC[0])))
```
